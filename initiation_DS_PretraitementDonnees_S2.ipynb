## **Les données ou valeurs manquantes**

```python
#Importer les données des vols commerciaux.
#Ces données contiennent des valeurs manquantes
flights_df = pd.read_csv("https://rcs.bu.edu/examples/python/DataAnalysis/flights.csv")
```

```python
#Vérifier l'existence des données manquantes
#Sélectionnez les lignes qui ont au moins une valeur manquante.
flights_df[flights_df.isnull().any(axis=1)].head()
```

```python
#Supprimer toutes les lignes contenant les valeurs manquantes
flights_df_na_dropped = flights_df.dropna()
flights_df_na_dropped[flights_df_na_dropped.isnull().any(axis=1)].shape[0]
```

```python
#Créer un nouveau DataFrame ne contenant que les lignes affectées
#par les valeurs manquantes
flights_df_new = flights_df[flights_df.isnull().any(axis=1)]
```

```python
#Remplacer les valeurs manquantes par zéro (0)
flights_df_new.fillna(0).head(10)
```

```python
#Remplacer les valeurs manquantes par la moyenne
flights_df_new.fillna(flights_df_new.mean(skipna=True, numeric_only=True)).head(10)
#Cette technique Ne marche pas si la colonne est "object" ou "string"
```
