## 2. Barres / Colonnes
**But analytique : Comparer des catégories (classement, top/bottom)**


### 3️⃣ Question métier
> Quel type de véhicule génère les sinistres **les plus coûteux** (coût moyen) ?

### 4️⃣ KPI (Indicateurs)
- **Coût moyen des sinistres** : CoûtTotal / NbSinistres (en €)
- **Fréquence (optionnel)** : NbSinistres / NbContrats (en %)

### 5️⃣ Visualisation choisie
Le **bar chart** est le plus efficace pour la **comparaison directe** entre catégories et le **classement**.


```python
# Données simulées (Assurance)
vehicules = ["Citadine", "Berline", "SUV", "Utilitaire", "Sportive", "Moto"]
cout_moyen = rng.normal(loc=[900, 1100, 1400, 1300, 2000, 800], scale=120).round(0)

df_bar = pd.DataFrame({"TypeVehicule": vehicules, "CoutMoyenSinistre": cout_moyen})
df_bar = df_bar.sort_values("CoutMoyenSinistre", ascending=True)
df_bar
```

```python
# Visualisation : Barres horizontales
fig, ax = plt.subplots()
ax.barh(df_bar["TypeVehicule"], df_bar["CoutMoyenSinistre"])

ax.set_title("Coût moyen des sinistres par type de véhicule (données simulées)")
ax.set_xlabel("Coût moyen (€)")
ax.set_ylabel("Type de véhicule")

for i, v in enumerate(df_bar["CoutMoyenSinistre"]):
    ax.text(v + 20, i, f"{int(v)}", va="center")
plt.show()
```

```python
# Interprétation automatique : top / bottom
top = df_bar.iloc[-1]
bottom = df_bar.iloc[0]
gap = top["CoutMoyenSinistre"] - bottom["CoutMoyenSinistre"]

print(f"Top coût moyen : {top['TypeVehicule']} = {top['CoutMoyenSinistre']:.0f} €")
print(f"Bottom coût moyen : {bottom['TypeVehicule']} = {bottom['CoutMoyenSinistre']:.0f} €")
print(f"Écart top-bottom : {gap:.0f} €")
```

### 6️⃣ Interprétation attendue
- Identifie la catégorie la plus coûteuse et estime l’écart avec la moins coûteuse.
- Discute une hypothèse métier (coûts de réparation, puissance, usage, exposition au risque).

### 7️⃣ Recommandation stratégique (exemples)
- **Tarification** : réévaluer les primes sur le segment le plus coûteux (avec contrôle conformité).
- **Prévention** : lancer une campagne ciblée (ex: conduite, stationnement, anti-vol).
- **Mesure** : suivre l’évolution du KPI (coût moyen) sur 3–6 mois après action.

## 3. Ligne / Aires
**But analytique : Montrer une tendance dans le temps (saisonnalité, ruptures)**


### 3️⃣ Question métier
> Le ratio **Sinistres/Primes (S/P)** se dégrade-t-il sur les 24 derniers mois ?

### 4️⃣ KPI (Indicateurs)
- **Ratio S/P** : Sinistres / Primes (cible : < 1.0 ou < 100%)
- **Variation mensuelle** : RatioSP(t) − RatioSP(t−1)

### 5️⃣ Visualisation choisie
La courbe est le standard pour les **séries temporelles**. On ajoute une ligne de seuil (100%) pour la décision.


```python
# Données simulées : série temporelle Ratio S/P
dates = pd.date_range("2024-01-01", periods=24, freq="MS")

trend = np.linspace(0.85, 1.05, len(dates))
season = 0.04 * np.sin(np.linspace(0, 3*np.pi, len(dates)))
noise = rng.normal(0, 0.015, len(dates))

df_line = pd.DataFrame({
    "Mois": dates,
    "Ratio_SP": (trend + season + noise).clip(0.6, 1.3)
})
df_line.head()
```

```python
# Visualisation : courbe + seuil 100%
fig, ax = plt.subplots()
ax.plot(df_line["Mois"], df_line["Ratio_SP"], marker="o")

ax.set_title("Évolution mensuelle du ratio Sinistres/Primes (S/P) — données simulées")
ax.set_xlabel("Mois")
ax.set_ylabel("Ratio S/P")

ax.axhline(1.0, linestyle="--", linewidth=1)
ax.text(df_line["Mois"].iloc[1], 1.01, "Seuil 100%", va="bottom")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Interprétation automatique : mois au-dessus du seuil
df_over = df_line[df_line["Ratio_SP"] > 1.0]
print(f"Nombre de mois avec S/P > 1 : {len(df_over)} / {len(df_line)}")
df_over.tail(8)
```

### 6️⃣ Interprétation attendue
- Repère les périodes où S/P dépasse 100% (non rentabilité).
- Attention : une tendance peut être influencée par la saisonnalité et les changements de portefeuille.

### 7️⃣ Recommandation stratégique (exemples)
- Mettre une **alerte** si S/P > 1 pendant ≥ 2 mois consécutifs.
- Segmenter l’analyse (âge, région, véhicule) avant d’ajuster le tarif.
- Tester une action (pilote/A-B) et mesurer l’effet sur S/P.

## 4. Barres empilées
**But analytique : Comparer la composition (partie/tout) entre groupes**


### 3️⃣ Question métier
> La **composition** des sinistres (matériel/corporel/vol) varie-t-elle selon la région ?

### 4️⃣ KPI (Indicateurs)
- **Volumes par type** : NbSinistres(type, région)
- **Part (%) par type** : NbSinistres(type, région) / Total(région)

### 5️⃣ Visualisation choisie
Empilées = composition. Version **100%** = comparer les parts.


```python
# Données simulées : sinistres par région et type
regions = ["Nord", "Sud", "Est", "Ouest", "Centre"]
types = ["Matériel", "Corporel", "Vol"]

df_stack = pd.DataFrame(
    rng.integers(50, 300, size=(len(regions), len(types))),
    columns=types, index=regions
)
df_stack
```

```python
# Empilées (valeurs absolues)
fig, ax = plt.subplots()
bottom = np.zeros(len(df_stack))

for t in df_stack.columns:
    ax.bar(df_stack.index, df_stack[t], bottom=bottom, label=t)
    bottom += df_stack[t].values

ax.set_title("Nombre de sinistres par région et type — données simulées")
ax.set_xlabel("Région")
ax.set_ylabel("Nombre de sinistres")
ax.legend(title="Type")

plt.tight_layout()
plt.show()
```

```python
# Empilées à 100% (parts)
df_stack_pct = df_stack.div(df_stack.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots()
bottom = np.zeros(len(df_stack_pct))
for t in df_stack_pct.columns:
    ax.bar(df_stack_pct.index, df_stack_pct[t], bottom=bottom, label=t)
    bottom += df_stack_pct[t].values

ax.set_title("Répartition (%) des sinistres par région et type — données simulées")
ax.set_xlabel("Région")
ax.set_ylabel("Part (%)")
ax.legend(title="Type")

plt.tight_layout()
plt.show()
```

```python
# Interprétation : région la plus touchée + type dominant
tot = df_stack.sum(axis=1).sort_values(ascending=False)
dominant_type = df_stack_pct.idxmax(axis=1)

print("Région avec le plus de sinistres :", tot.index[0], "=", int(tot.iloc[0]))
pd.DataFrame({"TypeDominant": dominant_type, "PartMax(%)": df_stack_pct.max(axis=1).round(1)}).sort_index()
```

### 6️⃣ Interprétation attendue
- Compare volumes vs parts : une région peut avoir peu de sinistres mais une **part Vol** élevée.
- Cherche un signal actionnable (territoire atypique).

### 7️⃣ Recommandation stratégique (exemples)
- Action territoriale (anti-vol) sur régions où Vol domine.
- Optimiser réseau de gestion (garages/expertise) si matériel domine.
- Mesurer : parts (%) + volumes avant/après.

## 5. Nuage de points (Scatter plot)
**But analytique : Analyser la relation entre deux variables numériques**


### 3️⃣ Question métier
> Quels contrats sont **non rentables** (Coût > Prime) ? Existe-t-il une relation Prime ↔ Coût ?

### 4️⃣ KPI (Indicateurs)
- **Prime annuelle** : € / an
- **Coût sinistres** : € / an
- **Ratio S/P** : Coût / Prime (cible < 1)

### 5️⃣ Visualisation choisie
Scatter = relation + outliers. Diagonale Coût=Prime pour repérer S/P>1.


```python
# Données simulées : Prime vs Coût + Ratio S/P
n = 250
prime = rng.normal(600, 180, n).clip(150, 1500)
cout = (0.9 * prime + rng.normal(0, 220, n) + rng.uniform(0, 250, n)).clip(0, 3000)
ratio = cout / prime

df_scatter = pd.DataFrame({"Prime": prime, "Cout": cout, "Ratio_SP": ratio})
df_scatter.head()
```

```python
# Scatter : taille ~ ratio S/P
fig, ax = plt.subplots()
sizes = (df_scatter["Ratio_SP"].clip(0, 3) * 30)

ax.scatter(df_scatter["Prime"], df_scatter["Cout"], s=sizes, alpha=0.6)
ax.set_title("Prime vs Coût sinistres (taille ~ ratio S/P) — données simulées")
ax.set_xlabel("Prime annuelle (€)")
ax.set_ylabel("Coût sinistres (€)")

xmin, xmax = df_scatter["Prime"].min(), df_scatter["Prime"].max()
ax.plot([xmin, xmax], [xmin, xmax], linestyle="--", linewidth=1)
ax.text(xmin + 50, xmin + 80, "Coût = Prime (S/P=1)", va="bottom")

plt.tight_layout()
plt.show()
```

```python
# Interprétation : part non rentables + top 10 S/P
non_rentables = df_scatter[df_scatter["Ratio_SP"] > 1.0]
print(f"S/P>1 : {len(non_rentables)} / {len(df_scatter)} = {len(non_rentables)/len(df_scatter):.1%}")

df_scatter.sort_values("Ratio_SP", ascending=False).head(10)
```

### 6️⃣ Interprétation attendue
- Les points au-dessus de la diagonale sont non rentables.
- Corrélation ≠ causalité : segmenter (âge/région/vehicule) avant décision.

### 7️⃣ Recommandation stratégique (exemples)
- Segmenter les non-rentables et cibler l’action (prévention/franchise/tarif).
- Piloter un test sur un sous-ensemble avant déploiement global.
- Mesurer : baisse de la part S/P>1.

## 7. Histogramme
**But analytique : Comprendre la distribution (asymétrie, seuils)**


### 3️⃣ Question métier
> Comment se **distribuent** les montants de sinistres ? Quelle part dépasse un seuil critique ?

### 4️⃣ KPI (Indicateurs)
- **Montant sinistre** : Sévérité (€)
- **Quantiles** : p50/p90/p95/p99

### 5️⃣ Visualisation choisie
Histogramme = distribution. Log-scale utile si queue longue.


```python
# Données simulées : montants de sinistres
n = 1200
montants = rng.lognormal(mean=7.2, sigma=0.6, size=n)
montants = np.clip(montants, 50, 50000)

df_hist = pd.DataFrame({"MontantSinistre": montants})
df_hist.describe(percentiles=[0.5, 0.9, 0.95, 0.99])
```

```python
# Histogramme
fig, ax = plt.subplots()
ax.hist(df_hist["MontantSinistre"], bins=40, alpha=0.85)
ax.set_title("Histogramme : montant des sinistres — données simulées")
ax.set_xlabel("Montant (€)")
ax.set_ylabel("Nombre de dossiers")
plt.tight_layout()
plt.show()
```

```python
# Variante log
fig, ax = plt.subplots()
ax.hist(df_hist["MontantSinistre"], bins=40, alpha=0.85)
ax.set_xscale("log")
ax.set_title("Histogramme (log) : montant des sinistres — données simulées")
ax.set_xlabel("Montant (€) [log]")
ax.set_ylabel("Nombre de dossiers")
plt.tight_layout()
plt.show()
```

```python
# Interprétation : proportion > seuil
seuil = 10000
prop = (df_hist["MontantSinistre"] > seuil).mean()
print(f"Proportion > {seuil}€ : {prop:.2%}")
```

## 8. Boxplot (boîte à moustaches)
**But analytique : Comparer distributions par groupe + détecter outliers**


### 3️⃣ Question métier
> Quels types de véhicules ont la **plus forte dispersion** et le plus d’outliers ?

### 4️⃣ KPI (Indicateurs)
- **Médiane** : valeur centrale
- **IQR** : Q3 − Q1 (dispersion)
- **Outliers** : valeurs extrêmes

### 5️⃣ Visualisation choisie
Boxplot = distribution + outliers + comparaison multi-groupes.


```python
# Données simulées : coûts par type (format long)
types = ["Citadine", "Berline", "SUV", "Utilitaire"]
data = []

for t in types:
    if t == "Citadine":
        vals = rng.lognormal(mean=6.9, sigma=0.55, size=250)
    elif t == "Berline":
        vals = rng.lognormal(mean=7.1, sigma=0.55, size=250)
    elif t == "SUV":
        vals = rng.lognormal(mean=7.25, sigma=0.6, size=250)
    else:
        vals = rng.lognormal(mean=7.15, sigma=0.65, size=250)
    vals = np.clip(vals, 80, 60000)
    data.append(vals)

df_box = pd.DataFrame({
    "TypeVehicule": np.repeat(types, 250),
    "CoutTotal": np.concatenate(data)
})
df_box.head()
```

```python
# Boxplot
fig, ax = plt.subplots()
series = [df_box.loc[df_box["TypeVehicule"] == t, "CoutTotal"] for t in types]
ax.boxplot(series, labels=types, showfliers=True)

ax.set_title("Boxplot : coût total par type de véhicule — données simulées")
ax.set_xlabel("Type de véhicule")
ax.set_ylabel("Coût total (€)")
plt.tight_layout()
plt.show()
```

```python
# Interprétation : médianes + IQR
med = df_box.groupby("TypeVehicule")["CoutTotal"].median().sort_values(ascending=False)

def iqr(x):
    return np.quantile(x, 0.75) - np.quantile(x, 0.25)

iqr_vals = df_box.groupby("TypeVehicule")["CoutTotal"].apply(iqr).sort_values(ascending=False)

print("Médianes (desc):")
print(med.round(0))
print("\nIQR (desc):")
print(iqr_vals.round(0))
```
